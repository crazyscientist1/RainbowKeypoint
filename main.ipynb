```python
import pandas as pd
import json
import numpy as np
from PIL import Image
```

```python
yTrue = [[], [], [], [], [], [], []]
xInp = [[], [], [], [], [], [], []]
```

```python
f = open(PATH + 'annotations.json')
data = json.load(f)

for i in data['metadata']:

    fileName = data['file'][data['metadata'][i]['vid']]['fname']

    image = Image.open(PATH + f'{fileName}')
    resize = (224,224)

    xCoord = int(data['metadata'][i]['xy'][1])

    yCoord = int(data['metadata'][i]['xy'][2])

    xScale = xCoord/np.shape(image)[1]
    yScale = yCoord/np.shape(image)[0]

    print("X: ", xCoord)
    print("Y: ", yCoord)
    print("X-scale: ", xScale)
    print("Y-scale: ", yScale)
    print("X-dim: ", np.shape(image)[0])
    print("Y-dim: ", np.shape(image)[1])

    image = image.resize(resize)
    xInp[int(data['metadata'][i]['vid'])-1] = image
    yTrue[int(data['metadata'][i]['vid'])-1].append([xScale * resize[0], yScale * resize[1]])
```

```python
f = open(PATH + 'rainbowProject1.json')
data = json.load(f)

for i in data['_via_img_metadata']:

    fileName = data['_via_img_metadata'][i]['filename']

    image = Image.open(PATH + f'{fileName}')


    yTrue.append([])
    for x in data['_via_img_metadata'][i]['regions']:
      xCoord = int(x['shape_attributes']['cy'])

      yCoord = int(x['shape_attributes']['cx'])
      yScale = xCoord/np.shape(image)[1]
      xScale = yCoord/np.shape(image)[0]

      print("X: ", xCoord)
      print("Y: ", yCoord)
      print("X-scale: ", xScale)
      print("Y-scale: ", yScale)
      print("X-dim: ", np.shape(image)[0])
      print("Y-dim: ", np.shape(image)[1])


      yTrue[len(yTrue) - 1].append([xScale * resize[0], yScale * resize[1]])

    xInp.append(image)
    image = image.resize(resize)
    resize = (224,224)
```

```python
InputShape = (size,size,3)
vggHead = tf.keras.applications.VGG16(include_top = False, weights = 'imagenet', input_shape = InputShape)
vggHead.trainable = False
```

```python
input_shape = (224, 224, 3)

inputs = Input(shape=input_shape)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Conv2D(32, (3, 3), activation='relu',)(x)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Conv2D(32, (3, 3), activation='relu',)(x)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Conv2D(32, (3, 3), activation='relu',)(x)
x = layers.MaxPooling2D((2, 2))(x)


x = layers.Flatten()(x)

# Dense layers
x = layers.Dense(2000, activation='relu')(x)
x = layers.Dense(100, activation='relu')(x)
x = layers.Dense(75, activation='relu')(x)
x = layers.Dense(50, activation='relu')(x)
x = layers.Dense(25, activation='relu')(x)
outputs = layers.Dense(14, activation='sigmoid')(x)

model = keras.Model(inputs, outputs, name="interestingCNN")
```

```python
x = vggHead.output

x = layers.Flatten()(x)

x = layers.Dense(4000, activation='relu')(x)
x = layers.Dense(1000, activation='relu')(x)
x = layers.Dense(705, activation='relu')(x)
x = layers.Dense(500, activation='relu')(x)
x = layers.Dense(25, activation='relu')(x)

# Add the final output layer
outputs = layers.Dense(14, activation='sigmoid')(x)

# Create the final model
model = Model(inputs=vggHead.input, outputs=outputs)

# Print the model summary
model.summary()
```

```python
optimizer = keras.optimizers.Adam()
batch_size= 30
epochSize = 600
epochs = 100
# train_dataset = tf.data.Dataset.from_tensor_slices((tf.constant(augmentedImages), tf.ragged.constant(augmentedCoords)))

# train_dataset = train_dataset.batch(batch_size)
```

```python
for epoch in range(epochs):
    print(f"\nStart of epoch: {epoch}")

    #Augmenting samples:
    epochImages = np.array([])
    epochCoords = []
    for sample in range(epochSize):
      image, coords = augmenter(xInp[sample % len(xInp)],size,yTrue[sample % len(xInp)])
      image = np.expand_dims(image, axis=0)
      if epochImages.size == 0:
        epochImages = image
      else:
        epochImages = np.vstack((epochImages,image))
      epochCoords.append(coords)

    train_dataset = tf.data.Dataset.from_tensor_slices((tf.constant(epochImages), tf.ragged.constant(epochCoords)))

    train_dataset = train_dataset.batch(batch_size)




    for step, (x_batch_train, y_batch_train) in enumerate(train_dataset):

        # Open a GradientTape to record the operations run
        # during the forward pass, which enables auto-differentiation.
        with tf.GradientTape() as tape:
            # Run the forward pass of the layer.
            # The operations that the layer applies
            # to its inputs are going to be recorded
            # on the GradientTape.

            logits = model(x_batch_train, training=True)  # Logits for this minibatch
            # Compute the loss value for this minibatch.
            loss_value = loss_func(logits, y_batch_train)
        # Use the gradient tape to automatically retrieve
        # the gradients of the trainable variables with respect to the loss.
        grads = tape.gradient(loss_value, model.trainable_weights)

        # Run one step of gradient descent by updating
        # the value of the variables to minimize the loss.
        optimizer.apply_gradients(zip(grads, model.trainable_weights))

    print("Epoch Ended")
    image = Image.open(PATH + 'rainbow.jpeg').resize((size,size))
    image = augmenter(image,size)

    predictions = model.predict(np.expand_dims(image, axis=0))
    image = Image.fromarray(np.uint8(image)).convert('RGB')
    draw = ImageDraw.Draw(image)
    for x,y in tf.reshape(predictions, [-1, 2]):
        x *= size
        y *= size
        draw.ellipse((x-5, y-5, x+5, y+5), fill='red', outline='red')
    display(image)
```
